### Data Wrangling Process

```python
# Check the shape of the DataFrame
print("Number of rows : ", df.shape[0])
print("Number of columns : ",df.shape[1])

print(df.shape)
```

```python
# print the column of the Dataframe 


print("Column Name : ", df.columns)
print("Column Name in List: ", df.columns.tolist())
```

```python
df.isnull().sum()
```

```python
# Fill missing values in the 'IMDb Rating' column with the mode (most frequent value)
# This ensures that missing IMDb ratings are replaced with a valid value

df['imdb_rating'].fillna(df['imdb_rating'].mode()[0],inplace=True)

# Fill missing values in the 'Year of Release' column with the mode (most frequent value)
# This ensures that missing release years are replaced with a valid value

df["release_year"].fillna(df['release_year'].mode()[0], inplace=True)

# Replace missing values in the 'Plot' column with 'No Data'
# This ensures that missing plot information is replaced with a placeholder

df["rent"].replace(np.nan, "No Data", inplace=True)
```

```python
# Check for missing values in each column of the cleaned DataFrame df_prime
# After cleaning, there should be no missing values in any column

cleaning_missing_values = df.isnull().sum()

# Print the number of missing values for each column after cleaning
print("Missing values after cleaning")
print(cleaning_missing_values)
```

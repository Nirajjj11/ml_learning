## Step 10: Data Transformation

Before we proceed to Bi-variate Analysis, Univariate analysis demonstrated the data pattern as some variables to be transformed.

Price and Kilometer-Driven variables are highly skewed and on a larger scale. Let’s do log transformation.

Log transformation can help in normalization, so this variable can maintain standard scale with other variables:

```python
# Function for log transformation of the column


def log_transform(data, cols):
      for col in cols:
            if (data[col] <= 0).any():   # check if 0 or negative exists
                  data[col + '_log'] = np.log1p(data[col])
            else:
                  data[col + '_log'] = np.log(data[col])
      data.info()
```

```python
log_transform(data,['Kilometers_Driven','Price'])
```

## Step 14: Impute Missing values

Missing data arise in almost all statistical analyses. There are many ways to impute missing values; we can impute the missing values by their Mean, median, most frequent, or zero values and use advanced imputation algorithms like KNN, Regularization, etc.

We cannot impute the data with a simple Mean/Median. We must need business knowledge or common insights about the data. If we have domain knowledge, it will add value to the imputation. Some data can be imputed on assumptions.

In our dataset, we have found there are missing values for many columns like Mileage, Power, and Seats.

We observed earlier some observations have zero Mileage. This looks like a data entry issue. We can fix this by filling null values with zero and then using the mean value of Mileage, since the mean and median values are nearly the same for this variable. Therefore, you choose the mean to impute the values.

```python
import numpy as np

# Replace 0 with NaN
data.loc[data["Mileage"] == 0.0, "Mileage"] = np.nan

# Check missing values
print(data["Mileage"].isnull().sum())

# Fill missing values with mean
data["Mileage"] = data["Mileage"].fillna(data["Mileage"].mean())

# Confirm again
print(data["Mileage"].isnull().sum())
```

```python
# Ensure numeric
data["Mileage"] = pd.to_numeric(data["Mileage"], errors="coerce")

# Replace 0 with NaN
data.loc[data["Mileage"] == 0, "Mileage"] = np.nan

# Fill with mean
data["Mileage"] = data["Mileage"].fillna(data["Mileage"].mean())
```

```python
def clean_mileage(df):
      df["Mileage"] = pd.to_numeric(df["Mileage"], errors="coerce")
      df.loc[df["Mileage"] == 0, "Mileage"] = np.nan
      df["Mileage"] = df["Mileage"].fillna(df["Mileage"].mean())
      return df

data = clean_mileage(data)

data['Mileage'].fillna(value=np.mean(data['Mileage']),inplace=True)
```

```python
import pandas as pd
import numpy as np

# -----------------------------
# 1️⃣ Clean Mileage
# -----------------------------

# Convert to numeric (if already numeric this won't harm)
data['Mileage'] = pd.to_numeric(data['Mileage'], errors='coerce')

# Replace 0 with NaN
data.loc[data['Mileage'] == 0, 'Mileage'] = np.nan

# Fill missing with mean
data['Mileage'] = data['Mileage'].fillna(data['Mileage'].mean())


# -----------------------------
# 2️⃣ Clean Engine
# -----------------------------

# Remove 'CC' if present
if data['Engine'].dtype == 'object':
      data['Engine'] = data['Engine'].str.replace(' CC', '', regex=False)

# Convert to numeric
data['Engine'] = pd.to_numeric(data['Engine'], errors='coerce')

# Fill missing using Brand + Model median
data['Engine'] = (
      data.groupby(['Brand', 'Model'])['Engine'].transform(lambda x: x.fillna(x.median()))
)

# Fallback (if group median still NaN)
data['Engine'] = data['Engine'].fillna(data['Engine'].median())


# -----------------------------
# 3️⃣ Clean Power
# -----------------------------

# Remove 'bhp' if present
if data['Power'].dtype == 'object':
      data['Power'] = data['Power'].str.replace(' bhp', '', regex=False)

# Convert to numeric
data['Power'] = pd.to_numeric(data['Power'], errors='coerce')

# Fill missing using Brand + Model median
data['Power'] = (
      data.groupby(['Brand', 'Model'])['Power'].transform(lambda x: x.fillna(x.median()))
)

# Fallback
data['Power'] = data['Power'].fillna(data['Power'].median())


# -----------------------------
# 4️⃣ Clean Seats
# -----------------------------

data['Seats'] = pd.to_numeric(data['Seats'], errors='coerce')

data['Seats'] = (
      data.groupby(['Brand', 'Model'])['Seats'].transform(lambda x: x.fillna(x.median()))
)

data['Seats'] = data['Seats'].fillna(data['Seats'].median())


# -----------------------------
# ✅ Final Check
# -----------------------------

print("Remaining Missing Values:\n")
print(data[['Mileage','Engine','Power','Seats']].isnull().sum())

print("\nData Types:\n")
print(data[['Mileage','Engine','Power','Seats']].dtypes)
```
